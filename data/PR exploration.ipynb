```python
def get_reviews(row):
    data = []
    reviews = row.reviews.values()
    for rev in reviews:
        data.append([row.name, row.created_by, rev["author"], rev["words_count"], rev["submitted_at"], rev["state"]])
    return data
```

```python
y = data.review_author
X = data.drop(["review_author", "words_count", "state", "response_seconds"], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,)

clf = KNeighborsClassifier(n_neighbors=1)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("KNN acc:", accuracy_score(y_test, y_pred))

clf = RandomForestClassifier(max_depth=20)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("RF acc:", accuracy_score(y_test, y_pred))
```

However, I am loosing valuable information from dropped features.
What is the solution?

### Neural Network embeddings!

```python
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
num_of_revs = len(data.review_author.unique())
num_of_revs
```

```python
y = keras.utils.to_categorical(data.review_author)
X = data.drop(["review_author"], axis=1)

## Too big values compared to other features
## by doing this accuracy is >0.9 (by not doing this the nn didn't even have 0.7)
X.response_seconds = X.response_seconds / X.response_seconds.max()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,)
```

```python
model = Sequential()

model.add(Dense(8))
model.add(Dense(100, activation="relu"))
model.add(Dense(num_of_revs, activation="softmax"))

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

model.fit(X_train, y_train, epochs=200)
```
